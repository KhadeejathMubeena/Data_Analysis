# DATA VISUALIZATION

```python
def data_label(a):
  for i in a.containers:
    a.bar_label(i,label_type='center',rotation=90)
```

```python
# df.groupby("Username")['Subscribers'].sum()
plt.figure(figsize = (10,5))
a = sns.barplot(x = 'Username',y = 'Subscribers',data = df.sort_values(by = "Subscribers",ascending = False).iloc[:5,:],color = 'lightgreen')
data_label(a)
plt.title('Subscribers by username')
plt.xticks(rotation=90)
plt.show()
```

**channels such as Mr Beast, T-Series, cocomelon-nursery Rhymes are having highest subscriber**

```python
df.info()
```

```python
df['Uploads'] = df["Uploads"].str.replace(',','')
df['Uploads'] = df['Uploads'].astype(int)
# df.info()
```

```python
df['Views'] = df["Views"].str.replace(',','')
df['Views'] = df['Views'].astype(int)

df.info()
```

```python
plt.figure(figsize = (10,5))
a = sns.barplot(x = 'Username',y = 'Views',data = df.sort_values(by = 'Views',ascending = False).iloc[:5,:],color = 'lightgreen')
data_label(a)
plt.title('Views by username')
plt.xticks(rotation=90)
plt.show()
```

**T-series, cocomelon-nursery Rhymes and set india are having highest views.**

```python
plt.figure(figsize = (10,5))
a = sns.barplot(x = 'Username',y = 'Uploads',data = df.sort_values(by = 'Uploads',ascending = False).iloc[:5,:],color = 'lightgreen')
data_label(a)
plt.title('Uploads by username')
plt.xticks(rotation=90)
plt.show()
```

**no.of uploads for ABP news, Aaj tak, ABS-CBN entertainment are more.**

```python
plt.figure(figsize=(15,9))
a=sns.countplot(x='Country',data=df,color='lightgreen')
data_label(a)
plt.title('Count of Channels by country',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

**India and USA are having highest channels compared to other countries.**

```python
df.sort_values(by='Views',ascending=False)
```

```python
df.info()
```

```python
df1=df.pivot_table(values=['Subscribers','Uploads','Views'],index='Country',aggfunc=['sum'])
plt.figure(figsize=(10,6))
a=sns.barplot(x='Country',y=('sum', 'Subscribers'),data=df1.sort_values(('sum', 'Subscribers'),ascending=False).iloc[:5,:],color='green')
data_label(a)
plt.title('Total Subscribers by country',fontsize=15)
plt.ylabel('Total subscribers')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
a=sns.barplot(x='Country',y=('sum', 'Uploads'),data=df1.sort_values(('sum', 'Uploads'),ascending=False).iloc[:5,:],color='green')
data_label(a)
plt.title('Total Uploads by country',fontsize=15)
plt.ylabel('Total Uploads')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
a=sns.barplot(x='Country',y=('sum', 'Views'),data=df1.sort_values(('sum', 'Views'),ascending=False).iloc[:5,:],color='green')
data_label(a)
plt.title('Total Views by country',fontsize=15)
plt.ylabel('Total Views')
plt.xticks(rotation=90)
plt.show()
```

**Total no.of subscribers, views are more for India and United states.**
